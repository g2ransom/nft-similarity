```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import pandas as pd
```

```python
import json
```

```python
with open("documents1.json", "r") as f:
    data = json.loads(f.read())
```

```python
data["data"][0]
```

```python
from pandas.io.json import json_normalize
```

```python
df = json_normalize(data, ["data"])
```

```python
nft_descriptions = df.description.array
```

```python
nft_descriptions[0:3]
```

```python
count_vectorizer = CountVectorizer()
vector_matrix = count_vectorizer.fit_transform(nft_descriptions)
vector_matrix
```

```python
tokens = count_vectorizer.get_feature_names()
```

```python
def create_dataframe(matrix, tokens):

    doc_names = ["doc_{}".format(i+1) for i, _ in enumerate(matrix)]
    df = pd.DataFrame(data=matrix, index=doc_names, columns=tokens)
    return(df)
```

```python
create_dataframe(vector_matrix.toarray(),tokens)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity_matrix = cosine_similarity(vector_matrix)
```

```python
cosine_similarity_matrix
```

```python
create_dataframe(cosine_similarity_matrix,["doc_{}".format(i+1) for i, _ in enumerate(vector_matrix.toarray())])
```
